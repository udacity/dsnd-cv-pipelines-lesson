```python
# imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 1. Data Acquision

```python
# load image
input_img = cv2.imread ('../../images/san_francisco_bay.jpg') 

# convert image to RGB and grayscale (matplotlib uses RGB, OpenCV uses BGR)
img_RGB = cv2.cvtColor(input_img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)
```

```python
# display image
plt.imshow(img_RGB)
plt.title(' San Francisco Bay')
```

**TODO**: Split the image in the three color channels and plot them separately in matplotlib. You can use OpenCV's split function - check the documentation if needed. Can you see from the images which color channel is which?

```python
# split the three color channels
r,g,b=cv2.split(img_RGB)

# plot original image, then the three channels for comparison
plt.imshow(img_RGB)
plt.title('original image')
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10, 30))
ax1.imshow(r,cmap = 'gray') 
ax1.title.set_text('red')
ax2.imshow(g,cmap = 'gray') 
ax2.title.set_text('green')
ax3.imshow(b,cmap = 'gray') 
ax3.title.set_text('blue')
```

```python
# we can make a cutout to work with a smaller part of the image (slicing)
cutout_RGB = img_RGB[0:600, 0:1600]
# cutout_RGB = img_RGB[0:img_RGB.shape[0], 0:img_RGB.shape[1]]
cutout_gray = cv2.cvtColor(cutout_RGB, cv2.COLOR_RGB2GRAY)
plt.imshow(cutout_gray, cmap='gray')
```

**TODO**: Flip the image horizontally and vertically and plot the results! Check out OpenCV's flip function for this task.

```python
# flip the image horizontally and vertically
flipped_0=cv2.flip(img_RGB,0) # vertical flip
flipped_1=cv2.flip(img_RGB,1) # horizontal flip

# display the original image and the two flipped versions
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10, 30))
ax1.imshow(img_RGB)
ax1.title.set_text('original image')
ax2.imshow(flipped_0)
ax2.title.set_text('vertical flip')
ax3.imshow(flipped_1)
ax3.title.set_text('horizontal flip')
```

**TODO**: Draw a red rectangle around the biggest island (Angel Island) in the original image. Check the OpenCV documentation for a suitable function to use!

```python
# draw a rectangle into the image
img_with_drawing = img_RGB.copy()
cv2.rectangle(img_with_drawing, (1000, 1000), (1500, 1500), (255, 0, 0), 10)
plt.imshow(img_with_drawing)
```

**TODO**: Resize the result to 400x400 pixels. Check the OpenCV documentation for a suitable function to use!

```python
# we can resize the image using different interpolation methods
resized_image = cv2.resize(img_with_drawing, (400, 400), interpolation=cv2.INTER_NEAREST)

# plot the resized image
plt.imshow(resized_image)
```
